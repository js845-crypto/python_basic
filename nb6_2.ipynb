```python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)


colors = ("red", "green", "blue")

for color in colors:
    print(color)



```

```python
nums = {10, 20, 30}

for num in nums: # set도 python에서 반복문이 되긴 한다. set자체는 원래 index가 없어서 반복문이 안되나 내부적으로 iterator 이용해서 되게 함
    print(num)
```

```python
student = {                          
    "name": "홍길동",
    "age": 20
}
# 키
for key in student:          #기본적으로 dictionary는 key값을 return함 
    print(key)
# 값
for value in student.values():
    print(value)

# 키 + 값
for key, value in student.items():    # items하면 튜플로 가져옴
    print(key, value)
```
