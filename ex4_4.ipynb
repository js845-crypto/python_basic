```python
a = 10
b = a

print(a, b) # 10 10
print(a is b)  # True 같은 객체 참조 #실질적 메모리 주소 비교
# 10 객체 메모리를 a와 b 둘다 동일하게 참조함, 공유 객체 참조. ( 두 가지 변수가 하나의 동일한 객체 메모리를 참조)
print(a==b) #값 비교
```

```python
a=10
b=10
print(a is b)
print(a == b)
#a=10, b=10이면 원래는 공유참조 아니나 메모리 낭비 때문에 공유참조로 처리함, 불변 객체에 한해서. 
#list면 가변 객체라서 공유참조 안하고도 참조할 수 있다.
```

```python
s1 = "Python"
s2 = s1

print(s1 is s2) # True
print(s1 == s2) # True
```

```python
s1="Python"
s2="Pytho"
s2="Pytho"+"n"
print(s1 is s2)
print(s1 == s2)
#s2에다가 "pytho"+"n"을 하면 문자열은 유일하기 때문에 동일한 객체라서 true true가 된다.
```

```python
s1= "python"
s2= "pytho"
s2= s2+"n"
print(s1 is s2)
print(s1 == s2)
# s2는 언제든지 바뀔 수 있기 때문에 s2+"n"은 새로운 객체를 생성해서 s2에 할당하는 것이기 때문에 s1과 s2는 다른 객체가 된다.
```

```python
nums1 = [10, 20, 30] #공유참조
nums2 = nums1

print(nums1 is nums2) # True
```

```python
nums1.append(40) 

print(nums1) # [10, 20, 30, 40]
print(nums2) # [10, 20, 30, 40]
```

```python
a = [1, 2, 3]
b = a

print(a == b) # True
print(a is b) # True

# 내장 함수 id()를 이용하면 동일 객체를 참조하는지 확인 가능
print(id(a))
print(id(b))
```
