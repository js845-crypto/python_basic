```python
# 함수를 변수에 저장
def hello():
    print("안녕하세요!")

# 함수를 변수에 저장
f = hello

# 변수로 함수 호출
f()
```

```python
# 함수를 다른 함수의 매개값(인수)로 전달
def hello():
    print("안녕하세요!")

def execute(func):
    func() # 함수 호출 연산자를 사용했다는 건 func 매개변수가 함수라는 의미

execute(hello)
```

```python
# 함수 정의는 함수선언문 이다 - 함수를 정의하는 문
# 람다는 함수 표현식 이다 - 람다 표현식의 결과로 함수를 반환한다.
# 람다 자체는 함수(객체)를 반환하고 람다를 호출하면 람다의 반환값이 반환된다.
# lambda 매개변수: 반환값

# 일반 함수 정의
def add1(a, b): # 람다와 다르게 함수는 이름이 있다 
    return a + b

# 람다 함수 , 위의 일반 함수와 동일한 람다이다. 이름이 없는 익명 함수
add2 = lambda a, b : a + b # add2에는 "lambda a, b : a + b" 람다가 반환한 함수(객체)가 대입

r1 = add1(5, 1)
print(r1)

r2 = add2(1, 3)
print(r2)
```

```python
# lambda는 익명함수라고 한다
r3= (lambda a, b : a + b)(10, 50)  # 파이썬에선 rambda에 바로 인수 대입해서 즉시 호출해서 사용하지 않음
print(r3)
```

```python
def square(x): # 함수를 중첩시킬 때마다 속 함수(fun에 해당하는 함수)를 일일히 정의해줘야한다
    return x * x

def test(fun):
    a = 2
    return fun(2) + 1000

r1 = test(square)
print(r1)
```

```python
def test(fun):
    a = 2
    return fun(2) + 1000

r4 = test(lambda x: x*1000) # 이렇게 하면 함수를 정의할 필요가 없어져서 편하다.
print(r4)
```

```python
def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True)

print(r)

r2 = sorted(s, key=score, reverse =False)
print(r2)

r3 = sorted(s, key=lambda e:e[1], reverse= False)
print(r3)
```

```python
# map() - 모든 요소를 변환
nums = [1, 2, 3, 4]

# 리스트의 각 요소(x)에 2를 곱한 결과를 반환
result = list(map(lambda x: x * 2, nums))

print(result)

result2 = map(lambda x: x * 2, nums) # map은 반복 가능한 객체 메모리 주소를 반환한다. list로 감싸서 리스트로 변환해야한다.
print(result2)
```

```python
# filter() - 조건(lambda)에 맞는 요소만 선택
nums = [1, 2, 3, 4, 5]

# 짝수인 요소만 선택
result = list(filter(lambda x: x % 2 == 0, nums)) # filter는 True, False를 기준으로 True인 데이터만 반환하여 남긴다.

print(result)

result2 = filter(lambda x: x % 2 == 0, nums) # filter는 반복 가능한 객체 메모리 주소를 반환한다. list로 감싸서 리스트로 변환해야한다.
print(result2)
```

```python
# sorted() - 정렬 기준(lambda) 지정

names = ["banana", "apple", "kiwi"]

# 문자열의 길이를 기준으로 오름차순 정렬
result = sorted(names, key=lambda x: len(x)) # sorted는 정렬된 리스트를 반환한다. 원본 리스트는 변경되지 않는다. list로 감싸지 않아도 된다
# sorted 함수는 reverse default 값이 False이다. 즉 오름차순 정렬이 기본이다
print(f"문자열의 길이를 기준으로 오름차순 정렬:{result}")
```

```python
#  stored()에 람다를 사용하면 데이터 과학에서 자주 사용되는 [dict, dict, ...] 자료구조를 원하는 키값으로 정렬이 가능하다.
students = [
    {"이름": "홍길동", "나이": 20, "점수": 90},
    {"이름": "이순신", "나이": 18, "점수": 85},
    {"이름": "김유신", "나이": 22, "점수": 95}
]

# "나이" 값을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["나이"])
print(f"나이 값을 기준으로 오름차순 정렬: {result}")

# x : 리스트의 각 딕셔너리
# x["나이"] : 각 딕셔너리의 "나이" 값을 꺼냄
# sorted()는 이 값을 기준으로 정렬한다.

# "점수"를 기준으로 내림차순 정렬
result2 = sorted(students, key=lambda x: x["점수"], reverse=True)
print(f"점수 기준 내림차순 정렬: {result2}")

result3 = sorted(students, key=lambda x: x["이름"])
print(f"이름 기준 오름차순 정렬: {result3}")
```

```python
# 여러 키를 기준으로 정렬도 가능하다
students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수 , 1순위가 같으면 2순위로 정렬한다
result = sorted(students, key=lambda x: (x["반"], x["점수"]))
print(result)
```

```python
# min() - 기준에 따라 최소값 찾기

names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))

print(result) # kiwi

```

```python
# max() - 기준에 따라 최대값 찾기

names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 긴 요소 반환
result = max(names, key=lambda x: len(x))

print(result) # banana
```

```python
#  any() - 하나라도 조건을 만족하는지 확인

nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환
r1 = map(lambda x: x % 2 == 0, nums)
print("r1 : ", r1) 
r2 = list(r1)
print("r2 : ", r2)
result = any(map(lambda x: x % 2 == 0, nums))   # any는 이전에 한 or 연산자와 동일하게 작동한다. 즉 앞에서 하나라도 True이면 break하고 바로 다음으로 넘어가서 속도가 빠르다. all은 속도가 느리다.
print(result) # True
```

```python
# all() - 모두 조건을 만족하는지 확인

nums = [2, 4, 6, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 모두 True이면 True 반환
result = all(map(lambda x: x % 2 == 0, nums))

print(result) #True
```
