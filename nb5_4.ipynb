```python
t= (2,4)
print(t)
print(type(t))
```

```python
t = 1, 3, 5 # 패킹 실행: 자동으로 튜플로(1,3,5) 묶어짐, list로 안됨
print(t) 
print(type(t))
l=list(t)
print(l,type(l))


x, y, z = t # 언패킹 실행 : 튜플이 자동으로 하나씩 풀림
print(x, y, z)
a, b = [3, 6] #(3,6) 튜플로 변경 후 언패킹 발생
print(a,b)
```

```python
#패킹 , 언패킹 시 *연산자
#  a, b = [ 1,2,3,4] too many values to unpack, 즉 갯수가 안맞아서 에러 나옴
*a, b = [1,2,3,4] # *연산자 사용하면 에러가 안나옴 
*c, d = (1,2,3,4) # 동일하게 list로 나와서 의미 없다 ()던 []던
print(a)
print(b)
print(c)
print(d)

a,*b = (1,2,3,4) # 이번엔 갯수가 1개 3개로 나뉨
print(a)
print(b)

e, *f , g = (1,2,3,4,) # *의 위치에 있는 곳을 하나로 묶는다 세개로 언패킹해야하는데 두번째 원소를 두개묶음으로
print(e)
print(f)
print(g)
# 갯수가 같아서 1:1로 하나씩 되도 *연산자 사용하면 list에 하나가 들어간다
```
