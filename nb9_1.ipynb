```python
if True:
    x = 10

print(x)        # 다른 언어와 다르게 if, for, try 등 블럭 안에서 선언한 변수도 밖에서 접근 가능. 즉 위에서 선언한 변수는 블럭에 상관없이 사용 가능 , 함수와 클래스는 제외


for i in range(3):
    value = i * 10

print(i)
print(value)

try:
    result = 100 / 10
except:
    pass

print(result)
```
