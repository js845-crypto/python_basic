```python
#구매금액
count = 0
total = 0
avg = 0
max_v = -1
min_v = 9999999

for o in orders:
    v = o["구매금액"]
    count += 1
    total = total+ v

    if v > max_v:
        max_v = v
    
    if v< min_v:
        min_v = v
    

avg = round(total / count , 2 )

print(f"행의 수 : {count}")

print ( f"구매금액합계: {total:,} 평균: {avg:,} 최대: {max_v:,} 최소: {min_v:,}")
```

```python
# 지역별 합계 , 지역벌 count

region_sum = {}
region_count= {}
region_avg = {}

for o in orders :
    r = o["지역"]
    v = o["구매금액"]

    if r not in region_sum: # 특정 지역 값이 없으면 (ex: 서울)  서울 : 0 값으로 추가
        region_sum[r] = 0    # region_sum ={"서울" : 0) 과 같다
        region_count[r] = 0 # region_count = {"서울" : 0}
    
    region_sum[r] = region_sum[r] + v    # 서울값 = 서울값 + 사용자의 구매금액
    region_count[r] +=  1

print(region_sum)
print(region_count)

for r in region_sum: # {'서울': 1880000, '부산': 498500.0, '대구': 0, '광주': 25000, '대전': 89000}
    print(f"{r} : {region_sum[r]/region_count[r]}")
```

```python
# 7. 등급별 평균
vip_sum = 0
vip_count = 0

normal_sum = 0
normal_count = 0

#level_sum = {"vip" : 0, "일반" : 0}
#level_count = {"vip": 0 , "일반" : 0} 이렇게 두개로 dictionary로 해도 된다 위의 4개 변수로 안하고. 변수 4개가 필요해지긴 함 
for o in orders:
    if o["회원등급"] == "vip":
        vip_sum = vip_sum + o["구매금액"]
        vip_count = vip_count + 1
    else:
        normal_sum = normal_sum + o["구매금액"]
        normal_count = normal_count + 1

print(f"vip: {vip_sum / vip_count:,}")
print(f"일반: {normal_sum / normal_count:,}")
```

```python
# 8. 0~2년, 3~5년, 6년 이상 각 그룹의 평균 구매금액을 구하시오.
price1 , price2, price3= 0, 0, 0 # 그룹별 구매금액 누적 price1, price2, price3 = ( 0 , 0, 0) 후 언패킹
count1, count2, count3 = 0, 0, 0 # 그룹별 카운트 누적


for o in orders:
    y = o["가입년수"]

    if y <= 2: # 2년이하
        price1 = price1 + o["구매금액"]
        count1 = count1 + 1

    elif y <= 5:
        price2 = price2 + o["구매금액"]
        count2 = count2 + 1

    else: # 6년이상
        price3 = price3 + o["구매금액"]
        count3 = count3 + 1

print(f"0~2년: {price1 / count1:,}")
print(f"3~5년: {price2 / count2:,}")
print(f"6~년: {price3 / count3:,}")
```
