```python
# 빈 딕셔너리
d1 = {} # set은 빈 set만들려면 set()
d2 = dict() # dictionary는 다른 컬렉션 type들과 다르게 index말고도 key가 있다
#순서 , 즉 index가 있어야 반복문이 가능함

print(d1)
print(d2)
```

```python
# {} 사용
student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}
print(student)
print(type(student))
```

```python
# dict() 사용
person = dict(name="김철수", age=30, city="서울")

print(person)
```

```python
# [()]를 dict으로 변환
data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]

fruit = dict(data)

print(fruit)
```

```python
#key는 불변객체만 가능
d = {"name": "홍길동"}   # 문자열 key
print(d)

d = {100: "학생"}  # 숫자로 key
print(d)

d = {       
    (1, 2): "좌표"          # 튜플로 key
}
print(d)


#d={
#   [1, 2]: "리스트"             # list는 가변객체라서 오류남
#}# error
#print(d)
```

```python
#연산

student = {"name": "홍길동"}
print(student)

# 요소조회
print(student["name"])

# 추가
student["age"] = 20
print(student)

# 수정
student["age"] = 30
print(student)

# 삭제
del student["age"]          #이건 사용하지말고 아래의 method 사용하자
print(student)

# in 연산
print("name" in student)
print("age" in student)        # 연산은 원본 유지하고 새로운 객체 생성한다 
```

```python
student = {"name": "홍길동"}

print(student.get("name"))
print(student.get("age")) # 없으면 None 반환

student = {"name": "홍길동", "age": 20}
print(student.keys())
print(student.values())
print(student.items())

student.update({"age": 30, "score": 90})
print(student)

age = student.pop("age")
print(age)
print(student)

student.clear()
print(student)
```
