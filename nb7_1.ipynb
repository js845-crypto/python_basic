```python
nums = []
for i in range(5): # 0 부터 4까지(5-1)
    nums.append(i)

print(nums)
```

```python
nums = range(1, 11)
list_nums = list(nums)
print(list_nums)

even = []

for i in list_nums:      # statement 구조
    pass  # 들여쓰기 tab 표시를 위해서 pass가 있음 python에
    if i%2 == 0:
        even.append(i)
print(even)
```

```python
# comprehension : for문에서 반환값 생기게 해서 변수에 할당할 수 있는 방법 , 식 구조

even2 = [ i for i in list_nums if i%2 ==0 ] 
# 위의 even.append(i) print(even) 인 짝수만 list한거랑 동일한 코드 dictionary로도 가능함, {key: } 이 형식으로 해야함, ai는 이런 식 형태를 선호함
print(even2)

print([i*10 for i in list_nums if i<=5])

names = ["홍길동", "이순신", "이황"]
print ([n+"님" for n in names ])

print ([n+"님" for n in names if n.startswith("이")])

# dictionary로 만드는 예제
names2 = ["홍길동", "이순신", "김유신", "박혁거세"] # n[:1], n[1:]
print({ n[:1] : n[1:]for n in names2})

```

```python
names2 = ["admin-1234", "guest 0000", "zeroday-7777"]  # - 문자를 토큰으로 이용하여 [0][1] 생성

print("admin-1234".split("-")[0], "admin-1234".split("-")[1])
```

```python
print({n.split("-")[0]:n.split("-")[1] for n in names2})
```

```python
for n in names2:
    print(n.split("-"))
```

```python
print({i:p for i, p in (n.split("-", ) for n in names2)})
```
