```python
num = int(input ("숫자를 입력하세요")) # 알파벳 입력 후 int로 전환하면 오류남
print(num*10)
```

```python
# try 문을 실행해서 error가 발생하면 except문을 실행함
try:
    num = int(input("숫자 입력 : ")) ** 2
    print(num)
except:
    print("숫자가 아닙니다.")

print("다음 실행문...")
```

```python
try:
    num = int(input("숫자 입력 : "))
    result = 10 / num
    print(result)
except ValueError:
    print("숫자를 입력하세요.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

print("종료")
```

```python
#예외 발생 여부와 관계 없이 "항상" 실행된다
try:
    print(10 / 2)
except:
    print("오류 발생")
finally:                # 예외가 발생해도 finally가 진행되므로 일반 코드와 다른 점이다! 즉 내가 예외처리 하지 못한 경우에 대해서도 일단 실행되게 함
    print("프로그램 종료")
```

```python
try:
    num = int(input("숫자 입력: "))
    result = 100 / num

except ValueError:
    print("숫자를 입력해야 합니다.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

else:
    print("계산 결과:", result)

finally:
    print("프로그램 종료")
```
