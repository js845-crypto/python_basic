```python
a = [10, 20, 30]
b = a #이것은 a가 가리키는 객체를 b도 가리키게 함, 즉 같은 객체를 가리킴(공동 참조)

print(a)
print(b)
print(a == b)
print(a is b)

b[0] = 999
print(a) 
print(b) 
```

```python
a = [10, 20, 30]
b = a[:]    #이것은 a의 모든 원소를 복사해서 새로운 객체를 만들어 b가 가리키게 함
# slicing은 새로운 객체를 만들어서 b가 가리키게 함, 즉 a와 b는 다른 객체를  가리킴
print(a)
print(b)
print(a == b) 
print(a is b)

b[0] = 999
print(a) # [10, 20, 30]
print(b) # [999, 20, 30]
```

```python
# append() : 맨 뒤에 1개 추가 , 메서드라서 원본 객체를 변경하는 것
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

```python
# insert() : 원하는 위치에 추가, 메서드라서 원본 객체를 변경하는 것
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
```

```python
# 정렬 메서드
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```

```python
nums = [50, 10, 40, 20]
n2 =nums.sort() # 오름차순
print(nums)
print(n2) # None, sort()는 원본 객체를 변경하는 메서드라서 반환값이 없음
```

```python
nums = [50, 10, 40, 20]
result = sorted(nums)
print(nums)
print(result)
```

```python
nums = [50, 10, 40, 20]
result = sorted(nums)
print(nums)
print(result)
nums = [50, 10, 40, 20]
result = sorted(nums, reverse=True)
print(result)

fruits = ["banana", "apple", "orange"]
result = sorted(fruits)
print(result)

# key매갲변수를 이용하여 문자열 길이로 정렬
fruits = ["watermelon", "kiwi", "banana", "apple"]
result = sorted(fruits, key=len)
print(result)

# 매개변수 key값으로는 함수가 온다. 뒤에서 람다함수를 배우면
# key값으로 람다식을 대입하여 다양한 요구 사항의 정렬도 가능하다.
# 자기 자신을 정렬할때는 sort()메서드, 다양한 형태의 정렬된 리스트를 반환할때는 sorted() 내장함수 사용
```
