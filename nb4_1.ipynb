```python
s= "python"
print(dir(s))
print(s.upper())
print(s) # 대부분의 객체 메서드는 자기자신은 수정하나 문자열은 불변이라 안바뀜
```

```python
text = "Python Programming"

print(text.upper())      # 모두 대문자
print(text.lower())      # 모두 소문자
print(text.capitalize()) # 첫 글자만 대문자
print(text.title())      # 각 단어 첫 글자 대문자
```

```python
text = "   Python   "

print(text.strip())   # 양쪽 공백 제거
print(text.lstrip())  # 왼쪽 공백 제거
print(text.rstrip())  # 오른쪽 공백 제거
#글자 중간 공백은 제거 못함 양끝만 가능
```

```python
text = "Python Programming"

print(text.find("Pro"))  #있으면 위치를 0부터시작해서 index 반환
print(text.find("Java")) #없으니 -1 반환
print(text.startswith("Python")) #python으로 시작하면 true
print(text.endswith("ing")) # inverse
```

```python
text = "I like Java"

print(text.replace("Java", "Python"))  #원본은 안바꾸는 것 그냥 반환만, 바꿀게 없으면 원본을 출력해서 보여줌
```

```python
data = "apple,banana,orange"

result = data.split(",") #문자열 목록들을 콤마를 기준으로 쪼개서 list로 만들음

print(result) # ['apple', 'banana', 'orange']
```

```python
fruits = ["apple", "banana", "orange"]  #list를 문자열로 합치기

result = ", ".join(fruits)

print(result)
```
