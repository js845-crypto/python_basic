```python
try:

    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
    print(10/0)   
finally:
    f.close()   # 직접 닫아야 함. 실수로 빼먹거나, 실행전 에러가 발생하면 메모리 누수가 발생한다.
    print("자원해지 성공")
```

```python
# with문을 사용하는 경우 : try except 하기 귀찮을 때, with 구문이 끝나면 에러가 발생하던 안하던 무조건 자원 해지 
with open("students.csv", "r", encoding="utf-8") as f:
    print(f.read())
    print(10/0)
    
```

```python
# 변수 만들어서 저장할 수도 있다
names = []
scores = []
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        names.append(name)
        scores.append(score)

print(names, scores)
```

```python
# 실제 업무에선 txt파일보다 csv파일을 읽는 일이 많을 것
# csv파일을 불러와서 사용할때는 open()함수 보다는 csv파일 전용 모듈 사용을 권장

import csv

# open()은 내장함수
with open("students.csv", "r", encoding="utf-8") as f:
    # reader() : csv 모듈 안에 있는 함수
    reader = csv.reader(f)
    print(type(reader))

    for row in reader:  # reader는 list를 가진 컬렉션 타입임을 알 수 있다
        print(row)
        print(type(row))
```

```python
# 헤더 (이름 ,국어, 영어) 항목 제거하고 두번째 줄 데이터부터 출력
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    next(reader) # 헤더 항목 건너뛰고 다음 것으로 넘어감, next를 사용할 수 있다는건 reader가 iterator 자료구조이다
# 시퀀스타입과 iterator 타입 찾아보기
    for row in reader:
        print(row)
```

```python
# DictReader()를 사용하여 첫번째 줄을 키로 하는 딕셔너리로 반환
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # csv.reader(f)로 안받고 dictionary로 바로 받아버림, 코드로 가공이 필요가 없음 , 헤드라인을 키로 받고 나머지를 값으로 받음

    for row in reader:
        print(row)
```

```python
# csv모듈의 reader반환 타입은 속도는 읽기 속도는 빠르지만(iterator), 자료구조 내부를 핸들링하는데는 여러가지 제한이 있다. 
# 필요하다면 list(reader)로 리스트로 만들어 인덱싱, 슬라이싱, len() 등을 사용하자.
# 큰 CSV 파일은 reader 반복으로 처리하고, 작은 CSV 파일은 list(reader)로 메모리에 올려 사용하자.
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)


print(data)
data[2]["영어"] = 80 # 전처리를 위해 None문자열로 저장된 csv 파일 항목을 수정이 가능해짐 list로 바꿧기 때문에.
print(data[2]) # 이전과 다르게 index로 접근이 가능해짐
```

```python
# csv.DictReader()로 읽어서 data를 수정했다면, csv.DictWriter() 또는 일반 파일 쓰기를 이용해 다시 저장하면 된다. 
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    
    # csv 파일 수정 작업을 하고 난 후에 ,, 여기선 수정은 안함
    
    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")    # csv 원본 파일이 수정되서 저장됨
```
