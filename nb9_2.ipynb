```python
def hello():
    """
    '안녕'을 출력하고 None 값을 반환하는 함수   
    """
    print("안녕")  #  트리플 문자열은 help()함수로 호출하면 함수에 대한 설명을 볼 수 있다. 또는 마우스 커서를 함수에 올려놓으면 함수에 대한 설명을 볼 수 있다.

result = hello()
# 함수호출은 표현식이다. 표현식은 값을 남긴다. 반환값이 없는 함수를 호출하면 None값을 남긴다.
print(result) # result에는 None값이 할당되어 있다.


help(hello)  # 아무것도 안 적으면 help함수로 호출해도 아무것도 안나옴
```

```python
# 자료 구조에 함수 저장하기
def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))


def add(x, y):
    return x + y

def sub(x, y):
    return x - y

calc = {
    "+": add,
    "-": sub
}

print(calc["+"](10, 5)) # calc dictionary의 "+" key에 해당하는 add함수를 호출
print(calc["-"](10, 5)) # calc dictionary의 "-" key에 해당하는 sub함수를 호출
```

```python
def calc(a, b):
    x = a + b
    y = a - b
    z = a * b
    return x , y, z # 3개의 값을 반환하는 것이 아니라 tuple로 묶어서 반환 , 왜냐하면 함수는 하나의 값만 반환할 수 있기 때문이다.

r1 = calc(1,2)
print(r1) # (3, -1, 2) 3개의 값이 tuple로 묶여서 반환된다.

r1, r2, r3 = calc(1,2) # tuple unpacking
print(r1)
print(r2)
print(r3)
```

```python
# 함수 호출시 "불변객체 인수 vs 가변객체 인수"를 사용했을때 차이
#  1) 불변 객체(Immutable)를 인수로 전달

# - int, float, str, tuple, bool, 등

def func(x):
    print("함수 안 (변경 전):", x, id(x))
    x += 10 # x는 불변객체라서 x가 가리키는 10에다가 10을 더하는게 아니고 10+10인 새로운 객체 생성한 후에 가리킨다
    
    print("함수 안:", x, id(x))

a = 10

func(a)

print("함수 밖:", a, id(a))

# 함수 안: 20 140714686129672
# 함수 밖: 10 140714686129352
# 함수 안의 x만 변경, 함수 밖의 a는 그대로
# id()값이 다르다


```

```python
# 2) 가변 객체(Mutable)를 인수로 전달 , 조심해야함
# - list, dict, set, 등

def func(nums):
    print("함수 안(변경 전):",nums, id(nums))
    nums.append(4) # 함수안의 변수 nums는 지역변수라서 함수 안에서만 접근 가능
    print("함수 안:",nums, id(nums))

data = [1, 2, 3]

func(data)

print("함수 밖:", data, id(data))

# 함수 안: [1, 2, 3, 4] 2514266949248
# 함수 밖: [1, 2, 3, 4] 2514266949248
# id()값이 동일하다.



```

```python
def func(nums):
    print("새로운 값 대입 전:", nums, id(nums))
    nums = [100, 200] # 대입은 오른쪽에서 왼쪽으로 진행되므로 [100,200] 생성 후에 nums가 새로운 객체를 가리키게 됨
    print("새로운 값 대입 후:", nums, id(nums))

data = [1, 2, 3]
func(data)
print(data, id(data))

# 새로운 값 대입 전: [1, 2, 3] 2514266872192
# 새로운 값 대입 후: [100, 200] 2514266954944
# [1, 2, 3] 2514266872192
```

```python
# 5. 함수 코드블록안에서 전역변수의 사용
# 전역변수는 문법적으로 사용할 수 있지만, 함수가 전역변수에 의존하면 재사용성과 유지보수성이 떨어진다. 
# 따라서 필요한 값을 매개변수로 전달하는 것이 가장 좋은 방법이며, 전역변수는 주로 상수나 프로그램 전체에서 공유하는 설정값의 읽기용도로만 사용하는 것이 좋다.
# 전역 변수 : 함수 정의와 같은 코드 레벨에 있는 변수
# 함수 코드블록 안에 전역변수를 사용하는 것은 좋지 않은 방법이다.  
# 대부분의 언어들은 함수안에 전역변수를 사용하는 것을 코딩 규칙으로 사용하지 않도록 한다.
# 왜 문법으로 막지 않느냐? "언어가 모든 것을 금지하기보다, 개발자가 책임 있게 사용하도록 한다"는 것이 프로그래밍 언어의 철학이다.

x = 10 

def fun_global():
    print(x) # x 전역변수를 사용

def fun_local():
    x = 333
    print(x) # 지역변수 x가 존재하기 때문에 지역변수 x를 사용


fun_global()
fun_local()
```

```python
# 6. 사용자 정의 함수 실습
#두 시퀀스의 교집합을 구하는 함수
# 위의 add 함수에서 반환 type을 명시하지 않았을 때 문자열 입력하면 문자열 더하기로 되서 자유로웠던 것을 상기
def my_intersect(a, b) -> list:   # -> list는 반환 type을 list라고 힌트를 주는 것, 다른 type을 반환시켜도 오류가 나지는 않음.
    """              
    a : 리스트
    b : 리스트
    반환 타입: 리스트
    내용 : 두개의 리스트의 중복되는 값만을 list로 반환
    """
    result = []
    for i in a:
        if i in b:
            result.append(i)
    return result
print(my_intersect([1,2,3], [1,5,6]))
a = [1,2,3,4,5,6,7,8]
b = [5,6,7,10,20,111]
print(my_intersect(a, b))
print(my_intersect({"id":"abc", "sn":"1111"}, {"id":"ayz", "pw":"1234"}))
```

```python
# 두 시퀀스의 공통 요소 개수

def my_count_common(a, b):
    count = 0
    for e in a:
        if e in b:
            count += 1
    return count


print(my_count_common([1, 2, 3], [2, 3, 5]))
print(my_count_common((1, 2, 3), (2, 3, 5)))
print(my_count_common({1, 2, 3}, {2, 3, 5}))
print(my_count_common({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# a에는 있고 b에는 없는 요소, 차집합 구하기

def my_difference(a, b):
    result = []
    for e in a:
        if e not in b: # 없는 것만 체크해서 result에 추가
            result.append(e)
    return result


print(my_difference([1, 2, 3], [2, 3, 5]))
print(my_difference((1, 2, 3), (2, 3, 5)))
print(my_difference({1, 2, 3}, {2, 3, 5}))
print(my_difference({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 두 시퀀스의 합집합

def my_union(a, b):
    result = []

    for e in a:
        if e not in result:
            result.append(e)

    for e in b:
        if e not in result:
            result.append(e)

    return result


print(my_union([1, 2, 3], [2, 3, 5]))
print(my_union((1, 2, 3), (2, 3, 5)))
print(my_union({1, 2, 3}, {2, 3, 5}))
```

```python
# 포함 여부 확인
# value가 시퀀스 안에 있는지 확인

def my_contains(data, value):

    for e in data:
        if e == value:
            return True # e 값이 있으면  True 값 반환하며 함수 종료

    return False


print(my_contains([1, 2, 3], 2))
print(my_contains(("a", "b", "c"), "d"))
print(my_contains({"id":"abc", "pw":"1234"}, "id"))
```

```python
# 특정 값이 몇 번 나오는지 계산

def my_count(data, target):

    count = 0

    for e in data:
        if e == target:
            count += 1

    return count


print(my_count([1, 2, 1, 3, 1], 1))
print(my_count(("a", "b", "a"), "a"))
print(my_count("banana", "a"))
```

```python
# 중복 제거

def my_unique(data):

    result = []

    for e in data:
        if e not in result:
            result.append(e)

    return result


print(my_unique([1, 2, 2, 3, 1, 4]))
print(my_unique(("A", "A", "B", "C")))
print(my_unique("banana"))
```
