```python
age = 20                                  # 반환 값이 필요하면
result = "성인" if age >= 19 else "미성년자"
print(result)
# 다른 언어에선 bool ? 참값 : 거짓값

print(f"당신은 {'성인' if age >= 19 else '미성년자'} 입니다") # ""안에 문자열은 ''를 사용하자
```

```python
age = 20
has_ticket = True            # none을 넣어도 된다. none도 False으로 간주, 비어있는 list도 False, 공백 문자열도 False


# 비교, 논리 즉 참 거짓을 판단할 수 있는 값을 반환하는 것을 넣어야함
if age >= 19 and has_ticket: # has_ticket은 참 거짓 판단 가능
    print("입장 가능")

else:
    print("입장 불가")
```
