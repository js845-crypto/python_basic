```python
a = {1,2,3,2} # list와 유사하지만 중복이 안되므로 코드가 간단해질 수 있어서 사용함
print(a) #원칙적으로는 합집합 교집합 차집합 연산 사용안하면 set 사용 안해야하지만 중복 관련 코드 알고리즘 싫기 때문에 사용 많이함

print(list(a)[0]) #set은 indexing 불가능 list로 바꿔야함
```

```python
h = {"1" : "admin", "pw": 1234 }
print(h, type(h))

b = {3, 4, 6}
print(b, type(b))

c = []
print(type(c))

d = {}
print(type(d))

e= set() #set은 dictionary랑 겹쳐서 함수 set을 사용해서 빈 set을 만든다
print(type(e))

```

```python
a = {1,2,3}
b = {3,4,5}

#교집합
print(a&b)
print(a.intersection(b))

#합집합
print(a|b)
print(a.union(b))

#차집합
print(a-b)
print(a.difference(b))
```

```python
s = {1,2,3}
s.add(4) #하나 추가
print(s)

s.update([5, 6]) #여러 개 추가
print(s)
s.remove(2) # 값 2 삭제 ( 없으면 오류)
print(s)
s.discard(10) # 값 없으면 그냥 무시
```

```python
s = {1,2,3}

print(len(s)) # 개수
print(2 in s) # 포함 여부

s.clear() #전체삭제
print(s) #set()
```
