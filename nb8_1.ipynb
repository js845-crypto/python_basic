```python
greet = "Hello"
while greet: # 문자열의 경우 공백이면 False, 아니면 True   , while에서 if로 갯수 바꾸는것보다 range로 하는게 속도가 더 빠르다
    print(greet)
    greet = greet[1:]
```

```python
a, b = 0, 10  
# 다중 할당은 실제로 존재하지 않음 내부적으로 오른쪽의 튜플 (0, 10)을 왼쪽의 변수들에 언패킹해서 각각 저장

while a < b:
    print(a)
    a = a + 1 #a += 1
```

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)

print("반복 종료")
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i) 
```

```python
if i % 2 == 0 :
    pass                # 코드 작성아직 완성 안됬을 때 빈 블럭 이면 오류나서 pass 키워드 사용해야함

for i in range(3):
    if i == 1:
        pass           
    print(i)
```

```python
# break가 있는 경우

for i in range(3):
    if i == 1:
        break
    print(i)
else:
    print("정상 종료") # break가 실행되었으므로 else는 실행되지 않는다.
```
