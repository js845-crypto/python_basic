```python
# 1. 정수 (int)
a = 10
print(a)
print(type(a))
```

```python
# 2. 부동 소수점 수 (float) : 오차를 가진다
b = 3.14
print(b)
print(type(b))
print(0.1+0.2)
```

```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))
print(3>5)
```

```python
# 5. 문자열 (str)
e = "Python"
print(e)
print(type(e))
print(e[1])
# e[1]="k" 문자열 객체 불변
```

```python
# 6. 리스트 (list)
f = [1, 2, 3]
print(f)
print(type(f))
print(f[1])
print(type(f[1]))  
f[0] = "abc" # 리스트 객체 가변
print(f)
```

```python
# 7. 튜플 (tuple)
g = (1, 2, 3)
print(g)
print(type(g))
print(g[1])
# g[1] = 9 튜플 불변 객체          불변 객체 사용 이유: 읽기 전용, 해시 가능 , 메모리 효율적 , 해킹 막기
```

```python
# 8. 딕셔너리 (dict)
h = {"id": "admin" , "pw" : "1234"} # admin이랑 1234가 값이고 id랑 pw는 key, key가 (1,3,5,6) 이런식으로 가능도 하다
print(h)
print(type(h))
print(h["id"])
h["pw"] = "0000" # 가변이다 값이
print(h) # 리스트랑 딕셔너리는 가변이다
```

```python
# 9. 집합 (set)
i = {1, 2, 3}            # 집합은 인덱스랑 키가 없다 그래서 print(i[0]) 이런식으로 접근 불가
print(i)
print(type(i))
# 집합은 중복된 값 넣어도 허용되지 않는다 print하면 중복 자동 제거
```

```python
# 10. 범위 (range)
j = range(5)
print(j)
print(type(j))
print(list(j)) # range는 리스트로 변환해야 값 확인 가능
print(type(list(j)))
```

```python
# 11. 바이트 (bytes)
k = b"abc"
print(k)
print(type(k))
```

```python
# 12. 바이트배열 (bytearray)
l = bytearray(b"abc")
print(l)
print(type(l))
```

```python
# 13. 없음 (NoneType)
m = None
print(m)
print(type(m))
```

```python
# 14. 함수 (function) 
def hello():             # 함수도 객체다
    print("Hello")
    print("bye")

print(hello) # 함수안에 저장된 주소값이 출력됨
print("hello")
hello()
print(type(hello))
hello
 

```

```python
# 15. 모듈 (module)
import math    # 모듈도 객체이며 파일이라고 생각

print(math)
print(type(math))
```

```python
# 16. 클래스 (type)
class Person:
    pass

print(Person)
print(type(Person))
```
