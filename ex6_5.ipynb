```python
students = [
    {"이름": "김철수", "성적": 95, "주민번호": "980101-1234567"},
    {"이름": "이영희", "성적": 88, "주민번호": "010305-4234567"},
    {"이름": "박민수", "성적": None, "주민번호": "950720-1234567"},  # 결측치
    {"이름": "최지은", "성적": 77, "주민번호": "030215-4234567"},
    {"이름": "정우성", "성적": 105, "주민번호": "990101-1234567"},   # 이상치
    {"이름": "한지민", "성적": 82, "주민번호": "020912-4234567"},
    {"이름": "강호동", "성적": -10, "주민번호": "700511-1234567"},   # 이상치
    {"이름": "신세경", "성적": 91, "주민번호": "971224-2234567"},
    {"이름": "유재석", "성적": 85, "주민번호": "720814-1234567"},
    {"이름": "송혜교", "성적": None, "주민번호": "811122-2234567"}, # 결측치
    {"이름": "차은우", "성적": 79, "주민번호": "970330-1234567"},
]

print(students) # print(students) 보다 students 출력하는 것이 Jupyter Notebook에서 가독성이 좋다
len(students)
```

```python
#전처리 -> 결측치 확인
missing_count = 0

for student in students:
    if student["성적"] is None: # None은 값이 아니라서 ==로하면 안됨 is로 해야함
        missing_count += 1

print("결측치 개수:", missing_count)
```

```python
#이상치 탐지
for student in students:
    score = student["성적"]

    if score is None or score < 0 or score > 100 :
        print(student)
```

```python
#이상치 처리
for student in students:

   for student in students:
    # 1. None 검사를 가장 첫 번째 'if'문으로 배치하여 에러를 원천 차단
    if student["성적"] is None:
        student["성적"] = None
        
    # 2. 이미 위에서 None이 걸러졌으므로 안전하게 크기 비교 가능
    elif student["성적"] > 100:
        student["성적"] = 100
        
    elif student["성적"] < 0:
        student["성적"] = 0
students
```

```python
# 결측치 처리
scores = []

for student in students:
    if student["성적"] is not None:
        scores.append(student["성적"])

avg = sum(scores) / len(scores)

for student in students:
    if student["성적"] is None:       # 결측치 값을 평균값으로 처리했다 여기선. 상황마다 domain마다 다르다
        student["성적"] = avg

students
```

```python
# 성별 구하기 (특성추출)
for student in students:

    gender_code = student["주민번호"].split("-")[1][0] # "-" 기준으로 두개의 list로 나눔

    if gender_code in ["1", "3"]:
        student["성별"] = "남"
    else:
        student["성별"] = "여"

students
```

```python
# 나이 구하기 (특성 추출)
from datetime import datetime

current_year = datetime.now().year

for student in students:
    rrn = student["주민번호"]
    yy = int(rrn[:2])
    code = rrn.split("-")[1][0]

    if code in ["1", "2"]:
        birth_year = 1900 + yy
    else:
        birth_year = 2000 + yy

    student["나이"] = current_year - birth_year

students
```

```python
# 여기서부터는 데이터 분석 기술통계
scores = []

for student in students:
    scores.append(student["성적"])

print("개수:", len(scores))
print("합계:", sum(scores))
print("평균:", sum(scores) / len(scores))
print("최대:", max(scores))
print("최소:", min(scores))
```

```python
# 성별 평균 성적
male_sum = 0
male_count = 0

female_sum = 0
female_count = 0

for student in students:

    if student["성별"] == "남":
        male_sum += student["성적"]
        male_count += 1

    else:
        female_sum += student["성적"]
        female_count += 1

print("남 평균:", male_sum / male_count)
print("여 평균:", female_sum / female_count)
```

```python
# 연령대별 평균 성적
age_group = {}

for student in students:

    decade = (student["나이"] // 10) * 10

    if decade not in age_group:
        age_group[decade] = []

    age_group[decade].append(student["성적"])

print(age_group)

for decade, scores in age_group.items():
    avg = sum(scores) / len(scores)
    print(f"{decade}대 평균: {avg:.2f}")
```
